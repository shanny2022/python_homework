For my final project, I am using the Global Superstore dataset. This week I cleaned the dataset, selected relevant business columns, filtered rows, added engineered features, and used groupby aggregations to find early business patterns.


```python
import glob
from pathlib import Path

import numpy as np
import pandas as pd


# Task 13: Final Project Week 6 Progress

def find_capstone_file():
    expected_columns = {"Sales", "Profit", "Category"}

    for file_path in glob.glob("/kaggle/input/**/*", recursive=True):
        path = Path(file_path)

        if path.suffix.lower() not in [".csv", ".xls", ".xlsx"]:
            continue

        try:
            if path.suffix.lower() == ".csv":
                preview = pd.read_csv(path, nrows=5)
            else:
                preview = pd.read_excel(path, nrows=5)
        except Exception:
            continue

        if expected_columns.issubset(set(preview.columns)):
            return file_path

    raise FileNotFoundError("Add the Global Superstore dataset to this Kaggle notebook.")


capstone_path = find_capstone_file()

if capstone_path.endswith(".csv"):
    superstore = pd.read_csv(capstone_path)
else:
    superstore = pd.read_excel(capstone_path)

superstore.columns = (
    superstore.columns
    .str.strip()
    .str.replace(" ", "_", regex=False)
    .str.replace("-", "_", regex=False)
)

print("Original data:")
display(superstore.head())
superstore.info()
```

```python
# Clean the dataset

superstore_clean = superstore.drop_duplicates().copy()

for column in superstore_clean.select_dtypes(include="object").columns:
    superstore_clean[column] = superstore_clean[column].astype(str).str.strip()

for column in ["Order_Date", "Ship_Date"]:
    if column in superstore_clean.columns:
        superstore_clean[column] = pd.to_datetime(superstore_clean[column], errors="coerce")

for column in ["Sales", "Profit", "Quantity", "Discount", "Shipping_Cost"]:
    if column in superstore_clean.columns:
        superstore_clean[column] = (
            superstore_clean[column]
            .astype(str)
            .str.replace("$", "", regex=False)
            .str.replace(",", "", regex=False)
        )
        superstore_clean[column] = pd.to_numeric(superstore_clean[column], errors="coerce")

critical_columns = [column for column in ["Sales", "Profit", "Order_Date"] if column in superstore_clean.columns]
superstore_clean = superstore_clean.dropna(subset=critical_columns)

print("Cleaned data:")
display(superstore_clean.head())
```

```python
# Filter, select, and transform

superstore_clean = superstore_clean.rename(columns={
    "Customer.ID": "Customer_ID",
    "Customer.Name": "Customer_Name",
    "Order.Date": "Order_Date",
    "Order.ID": "Order_ID",
    "Order.Priority": "Order_Priority",
    "Product.ID": "Product_ID",
    "Product.Name": "Product_Name",
    "Row.ID": "Row_ID",
    "Ship.Date": "Ship_Date",
    "Ship.Mode": "Ship_Mode",
    "Shipping.Cost": "Shipping_Cost",
    "Sub.Category": "Sub_Category",
})

subcategory_column = "Sub_Category"
geo_column = "Market"
customer_column = "Customer_Name"

selected_columns = [
    "Order_Date",
    "Ship_Date",
    "Category",
    subcategory_column,
    geo_column,
    customer_column,
    "Sales",
    "Profit",
    "Quantity",
    "Discount",
    "Shipping_Cost",
]

superstore_analysis = superstore_clean.loc[
    superstore_clean["Sales"] > 0,
    selected_columns
].copy()

superstore_analysis["Category"] = superstore_analysis["Category"].str.title()
superstore_analysis[subcategory_column] = superstore_analysis[subcategory_column].str.title()
superstore_analysis[geo_column] = superstore_analysis[geo_column].str.title()

display(superstore_analysis.head())
```

```python
# Feature engineering

superstore_analysis["Order_Date"] = pd.to_datetime(
    superstore_analysis["Order_Date"],
    errors="coerce"
)

if "Ship_Date" in superstore_analysis.columns:
    superstore_analysis["Ship_Date"] = pd.to_datetime(
        superstore_analysis["Ship_Date"],
        errors="coerce"
    )

superstore_analysis = superstore_analysis.dropna(subset=["Order_Date"]).copy()

superstore_analysis["Order_Year"] = superstore_analysis["Order_Date"].dt.year
superstore_analysis["Order_Month"] = superstore_analysis["Order_Date"].dt.month
superstore_analysis["Gross_Margin"] = superstore_analysis["Profit"] / superstore_analysis["Sales"]

superstore_analysis["Profit_Status"] = superstore_analysis["Profit"].map(
    lambda profit: "Profitable" if profit > 0 else "Unprofitable"
)

if "Discount" in superstore_analysis.columns:
    superstore_analysis["Discount_Bucket"] = pd.cut(
        superstore_analysis["Discount"],
        bins=[-0.01, 0, 0.10, 0.30, 1.00],
        labels=["None", "Low", "Medium", "High"],
    )

if "Ship_Date" in superstore_analysis.columns:
    superstore_analysis["Shipping_Days"] = (
        superstore_analysis["Ship_Date"] - superstore_analysis["Order_Date"]
    ).dt.days

display(superstore_analysis.head())
```

```python
# Aggregation 1: Product performance

category_summary = (
    superstore_analysis
    .groupby(["Category", subcategory_column])
    .agg(
        total_sales=("Sales", "sum"),
        total_profit=("Profit", "sum"),
        average_profit=("Profit", "mean"),
        average_margin=("Gross_Margin", "mean"),
        order_count=("Profit", "count"),
    )
    .reset_index()
    .sort_values("total_profit", ascending=False)
)

print("Most profitable product groups:")
display(category_summary.head(10))

print("Least profitable product groups:")
display(category_summary.tail(10))
```

```python
# Aggregation 2: Geographic performance

geo_summary = (
    superstore_analysis
    .groupby(geo_column)
    .agg(
        total_sales=("Sales", "sum"),
        total_profit=("Profit", "sum"),
        average_margin=("Gross_Margin", "mean"),
        order_count=("Profit", "count"),
    )
    .reset_index()
    .sort_values("total_profit", ascending=False)
)

display(geo_summary)
```

```python
# Aggregation 3: Customer lifetime value

customer_value = (
    superstore_analysis
    .groupby(customer_column)
    .agg(
        total_sales=("Sales", "sum"),
        total_profit=("Profit", "sum"),
        order_count=("Profit", "count"),
    )
    .reset_index()
    .sort_values("total_profit", ascending=False)
)

display(customer_value.head(10))
```

```python
# Aggregation 4: Discount impact

if "Discount_Bucket" in superstore_analysis.columns:
    discount_summary = (
        superstore_analysis
        .groupby("Discount_Bucket", observed=False)
        .agg(
            total_sales=("Sales", "sum"),
            total_profit=("Profit", "sum"),
            average_discount=("Discount", "mean"),
            average_margin=("Gross_Margin", "mean"),
            order_count=("Profit", "count"),
        )
        .reset_index()
    )

    display(discount_summary)
```

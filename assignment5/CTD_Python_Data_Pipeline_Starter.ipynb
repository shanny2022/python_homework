```python
import os
import re
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", None)


def list_input_files():
    root = Path("/kaggle/input")
    if not root.exists():
        print("No Kaggle input directory found yet. Add one of the capstone datasets in Kaggle.")
        return []
    files = sorted(path for path in root.rglob("*") if path.is_file())
    for path in files:
        print(path)
    return files


input_files = list_input_files()
```

## Capstone Dataset Review

This starter notebook is for the Kaggle Data Pipeline capstone project. I will use it to inspect the available datasets, choose one, clean the data, create useful features, aggregate the results, and build charts that support at least three insights.

```python
csv_files = [path for path in input_files if path.suffix.lower() == ".csv"]
print(f"CSV files found: {len(csv_files)}")

previews = {}
for path in csv_files:
    try:
        sample = pd.read_csv(path, nrows=5)
    except Exception:
        try:
            sample = pd.read_csv(path, sep=";", nrows=5)
        except Exception as error:
            print(f"Could not preview {path}: {error}")
            continue
    previews[str(path)] = sample
    print("\n", path)
    display(sample)
```

## Dataset Choice

I am reviewing the available datasets first. My current plan is to choose the dataset with clear numeric columns, date columns, and categories so I can create meaningful cleaning steps, aggregations, and visualizations.

```python
# After choosing a dataset, update this path to the CSV you want to analyze.
selected_path = csv_files[0] if csv_files else None

if selected_path is not None:
    df = pd.read_csv(selected_path)
    print(f"Loaded: {selected_path}")
    display(df.head())
    df.info()
    print("\nMissing values:")
    print(df.isna().sum().sort_values(ascending=False).head(20))
    print("\nDuplicate rows:", df.duplicated().sum())
```

## Initial Cleaning Notes

Cleaning steps to document here:

- Remove exact duplicate rows if they do not represent real repeated observations.
- Convert date columns with `pd.to_datetime`.
- Convert currency or percent columns to numeric values.
- Standardize text columns with `.str.strip()` and consistent casing.
- Fill or remove missing values depending on whether the column is required for analysis.

```python
if selected_path is not None:
    df_clean = df.copy()
    df_clean = df_clean.drop_duplicates().reset_index(drop=True)

    for column in df_clean.select_dtypes(include="object").columns:
        df_clean[column] = df_clean[column].astype(str).str.strip()

    print("Cleaned shape:", df_clean.shape)
    display(df_clean.head())
```
